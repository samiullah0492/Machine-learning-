Pandas has two main data structures:

### Series
A Series is a one-dimensional array with indexes. It stores a single column or row of data in a DataFrame and can hold any data type.

### DataFrame

A Series is like a column, while a DataFrame is the whole table.

## Slicing with `loc`

```python
import pandas as pd

# Create a sample DataFrame
data = {
    "Name": ["Ali", "Sara", "Bilal", "Hina"],
    "Marks": [85, 92, 76, 89],
    "Age": [20, 21, 19, 22]
}

df = pd.DataFrame(data)
print("Original DataFrame:\n", df)
```

```python
# 1. Select row with label 2
print(df.loc[2])
```

```python
# 2. Select specific rows and columns by name
print(df.loc[1:3, ["Name", "Marks"]])
```

```python

# 3. Apply a filter: students with Marks > 85
print(df.loc[df["Marks"] > 85])
```

## Slicing with `iloc`

```python
df
```

```python
# 1. Select row at index 0
print(df.iloc[0])
```

```python
# 2. Select 2 rows and 2 columns by position
print(df.iloc[0:2, 0:2])
```

```python
# 3. Select last row using negative index
print(df.iloc[-1])
```

# GroupBy
The `groupby` method allows you to group rows of data and call aggregate functions.

```python
data = {'Company':['GOOG','GOOG','MSFT','MSFT','FB','FB'],
       'Person':['Sam','Charlie','Amy','Vanessa','Carl','Sarah'],
       'Sales':[200,120,340,124,243,350]}
```

```python
df = pd.DataFrame(data)
```

```python
df
```

### Using the `.groupby()` Method
You can use `.groupby()` to group rows based on a column name. For example, grouping by `Company` creates a `DataFrameGroupBy` object.

```python
df.groupby('Company')
```

##### Save the GroupBy object as a new variable:

```python
by_comp = df.groupby("Company")
type(by_comp)
```

#### Call aggregate methods on the GroupBy object:

```python
by_comp.std(numeric_only=True)
```

```python
df.groupby('Company').std(numeric_only=True)
```

```python
by_comp.std(numeric_only=True)
```

```python
by_comp.min()
```

```python
by_comp.max()
```

```python
by_comp.count()
```

```python
by_comp.describe()
```
